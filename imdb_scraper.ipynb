```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
url = ("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
html = requests.get(url).text
soup = BeautifulSoup(html, "html5lib")
movie_list = soup('tbody', {'class' : 'lister-list'})

movies = [movie('tr') for movie in movie_list]

movie_num = len(movies[0])

titles = [movies[0][i]('td', {'class', 'titleColumn'})[0].contents[1].text for i in range(movie_num)] 
years = [movies[0][i]('td', {'class', 'titleColumn'})[0]('span')[0].contents[0][1:-1] for i in range(movie_num)] 
ratings = [float(movies[0][i]('td', {'class', 'ratingColumn imdbRating'})[0].contents[1].text) for i in range(movie_num)]
```

```python
df = pd.DataFrame()
df = df.assign(title=titles, year=years, rating=ratings)
df
```

```python
df.to_csv("top250.csv")
```
